```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='DEATH_EVENT')
plt.title('Survival (0) vs Death (1) Distribution')
plt.show()
```

```python
num_cols = ['age', 'creatinine_phosphokinase', 'ejection_fraction',
            'platelets', 'serum_creatinine', 'serum_sodium', 'time']

plt.figure(figsize=(14,10))
for i, col in enumerate(num_cols):
    plt.subplot(3,3,i+1)
    sns.boxplot(data=df, x='DEATH_EVENT', y=col)
plt.tight_layout()
plt.show()
```

```python
# sns dist plot
for i, col in enumerate(num_cols):
    sns.distplot(df[col])
    plt.show()
```

```python
# log transform
df['serum_creatinine'] = np.log1p(df['serum_creatinine'])
df['creatinine_phosphokinase'] = np.log1p(df['creatinine_phosphokinase'])
# plot again
for i, col in enumerate(num_cols):
    sns.distplot(df[col])
    plt.show()
```

```python
cat_cols = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']
plt.figure(figsize=(16,4))
for i, col in enumerate(cat_cols):
    plt.subplot(1,5,i+1)
    sns.countplot(data=df, x=col, hue='DEATH_EVENT', palette='viridis')
plt.tight_layout()
plt.show()
```

```python
df.isnull().sum()
```

```python
# corr
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols + ['DEATH_EVENT']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# scale
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
x = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']
```

```python
# split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# simple logreg
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# class weight
logreg = LogisticRegression(class_weight='balanced')
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, roc_auc_score
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB()
}
```

```python
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'roc_auc': roc_auc_score(y_test, y_pred)
    }
    print(f"\n{name} Performance:")
    print(classification_report(y_test, y_pred))

    ConfusionMatrixDisplay.from_estimator(
        model, X_test, y_test,
        display_labels=['Survived', 'Died']
    )
    plt.title(f'{name} Confusion Matrix')
    plt.show()
```

```python
# Compare ROC-AUC
pd.DataFrame.from_dict(results, orient='index')['roc_auc'].sort_values().plot(kind='barh')
plt.title('Model Comparison (ROC-AUC)')
plt.show()
```

```python
# grid search
from sklearn.model_selection import GridSearchCV
param = {'C': [0.1, 1, 10, 100],
         'gamma': [10, 1, 0.1, 0.01,],
         'kernel': ['rbf', 'poly','linear']}
grid_search_svm = GridSearchCV(SVC(), param, refit=True, verbose=3)
grid_search_svm.fit(X_train, y_train)
print(grid_search_svm.best_params_)
print(grid_search_svm.best_estimator_)
```

```python
y_pred = grid_search_svm.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(roc_auc_score(y_test, y_pred))
```

```python
# cross validation on diffrent models
from sklearn.model_selection import cross_val_score
models = {
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB()
}
for name, model in models.items():
    scores = cross_val_score(model, x, y, cv=5)
    print(f"{name} Cross-Validation Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
```

```python
# Handle imbalanced data: Using SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

print("Shape of training data after SMOTE:", X_resampled.shape)
print("Distribution of target variable after SMOTE:\n", y_resampled.value_counts())
```

```python
# Retrain models with resampled data
models_resampled = {
    'Logistic Regression (SMOTE)': LogisticRegression(),
    'K-Nearest Neighbors (SMOTE)': KNeighborsClassifier(),
    'Support Vector Machine (SMOTE)': SVC(),
    'Decision Tree (SMOTE)': DecisionTreeClassifier(),
    'Naive Bayes (SMOTE)': GaussianNB()
}

results_resampled = {}
for name, model in models_resampled.items():
    model.fit(X_resampled, y_resampled)
    y_pred = model.predict(X_test)
    roc_auc = roc_auc_score(y_test, y_pred)


    results_resampled[name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'roc_auc': roc_auc
    }
    print(f"\n{name} Performance:")
    print(classification_report(y_test, y_pred))

    ConfusionMatrixDisplay.from_estimator(
        model, X_test, y_test,
        display_labels=['Survived', 'Died']
    )
    plt.title(f'{name} Confusion Matrix')
    plt.show()
```

```python
# compare roc_auc_score
pd.DataFrame.from_dict(results_resampled, orient='index')['roc_auc'].sort_values().plot(kind='barh')
plt.title('Model Comparison (ROC-AUC)')
plt.show()
```

```python
param_smote = {'C': [0.1, 1, 10, 100],
               'gamma': [10, 1, 0.1, 0.01],
               'kernel': ['rbf', 'poly', 'linear']}

grid_search_svm_smote = GridSearchCV(SVC(), param_smote, refit=True, verbose=3,scoring = 'accuracy')
grid_search_svm_smote.fit(X_resampled, y_resampled)

print("\nBest parameters for SVM after SMOTE:", grid_search_svm_smote.best_params_)
print("Best estimator for SVM after SMOTE:", grid_search_svm_smote.best_estimator_)
```

```python
y_pred = grid_search_svm_smote.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# cross validation on resampled
for name, model in models_resampled.items():
    scores = cross_val_score(model, x, y, cv=5)
    print(f"{name} Cross-Validation Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
```
